## <span  style="color:Green">**<center>let's take a gentle dip into the  Functional Programming.**</center></span>



<p>&#9658; Functional programming, is a coding paradigm in which the building blocks are immutable values and “pure functions” that share no state with other functions.</p> 
<p>&#9658; Every time a pure function has a given input, it will return the same output – without mutating data or causing side effects.</p>
<p>&#9658; In this sense, pure functions are often compared to mathematical operations.</p>
<p>&#9658; For example, 3 plus 4 will always equal 7, regardless of what other mathematical operations are being done, or how many times you’ve added things together before.</p>



</p>&#9658; Although functional programming has existed since the 1950s, and is implemented by a long lineage of languages, it doesn’t fully describe a programming language. </p>
<p> &#9658;  Clojure, Common Lisp, Haskell, and OCaml are all functional-first languages with different stances on other programming language concepts, like the type system and strict or lazy evaluation. </p>

```python
from IPython.display import display, Markdown
display(Markdown("# (note, in purely functional languages they are not called variables)!"))
# Create a global variable `A`.
A = 5

def impure_sum(b):
    # Adds two numbers, but uses the
    # global `A` variable.
    return b + A

def pure_sum(a, b):
    # Adds two numbers, using
    # ONLY the local function inputs.
    return a + b

print(impure_sum(6))

print(pure_sum(4, 6))

```

 ## <span  style="color:red">Typically, in functional programming, we do not use loops. We use recursion. Recursion is a mathematical concept, usually, it means “feeding into itself”. With a recursive function, the function repeatedly calls itself as a sub-function. Here’s a nice example of a recursive function in Python:</span>

```python
def factorial_recursive(n):
    # Base case: 1! = 1
    if n == 1:
        return 1

    # Recursive case: n! = n * (n-1)!
    else:
        return n * factorial_recursive(n-1)
factorial_recursive(3)
```

 # The Map Function
 
 The map() function takes in an iterable (ie. list), and creates a new iterable object, a special map object. The new object has the first-class function applied to every element.

```python
x = [1, 2, 3, 4, 5]
print(list(map(lambda num: num * num, x)))
```

# The Reduce Function
Reduce is a function that turns an iterable into one thing. Typically you perform a computation on a list to reduce it down to one number. Reduce looks like this:

```python
from functools import reduce

product = reduce((lambda x, y: x * y),[1, 2, 3, 4])
product
```

# The filter function
The filter function takes an iterable and filters out all the things you don’t want in that iterable.

```python
x = range(-5, 5)
all_less_than_zero = list(filter(lambda num: num < 0, x))
all_less_than_zero
```

# Higher order functions
Higher order functions can take functions as parameters and return functions. 

All functions in Python are first class objects. A first class object is defined as having one or more of these features:

<p>&#9658;Created at runtime</p>
<p>&#9658;Assigned tro a variable or element in a data structure</p>
<p>&#9658;Passed as an argument to a function</p>
<p>&#9658;Returned as the result of a function</p>
<p>&#9658;So all functions in Python are first class and can be used as a higher order function.</p>

```python
def summation(nums):
    return sum(nums)

def action(func, numbers):
    return func(numbers)

print(action(summation, [1, 2, 3,4,5]))
```

# Partial application
Partial application (also called closures) is a bit weird, but are super cool. You can call a function without supplying all the arguments it requires. Let’s see this in an example. We want to create a function which takes 2 arguments, a base and an exponent, and returns base to the power of the exponent, like so:

```python
from functools import partial

def power(base, exponent):
    return base ** exponent
square = partial(power, exponent=2)
print(square(2))
```

# Find the greatest common divisor of a list of numbers using Reduce. Explain how the code works as well. Explain how it works as well and how it relates to functional programing.

```python
num=(20,30,40,5)
def gcd(a,b):
    return a if not b else gcd(b, a%b)
res=reduce(lambda m,n: m if not n else gcd(n,m%n),num)
print(res)
```
